```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats

page_speed = np.random.normal(10, 4, 1000)
page_request = 120 - (page_speed * 2) + np.random.normal(0, 1, 1000)
plt.scatter(page_speed, page_request);
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(page_speed, page_request)
```

```python
def predict(x):
    return slope * x + intercept
    
predicted_request = predict(page_speed)
plt.scatter(page_speed, page_request);
plt.plot(page_speed, predicted_request, color='red')
plt.show();

r_squared = r_value ** 2 

```
