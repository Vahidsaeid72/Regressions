```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats

np.random.seed(2)
page_speed = np.random.normal(3.0, 1.0, 1000)
page_request = np.random.normal(100, 2, 1000)**2/ page_speed
plt.scatter(page_speed, page_request);
x = np.array(page_speed)
y = np.array(page_request)

```

```python
p4 = np.poly1d(np.polyfit(x, y, 4))
print(p4)
xp = np.linspace(0, 7, 100)
plt.scatter(page_speed, page_request);
plt.plot(xp, p4(xp), "r")
plt.show();
```
